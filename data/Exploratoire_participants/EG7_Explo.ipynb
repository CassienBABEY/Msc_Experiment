```python
import pandas as pd
import numpy as np
```

```python
EG7 = pd.read_csv("data_expe/EG7_data.csv")
EG7
```

```python
EG7 = EG7[['anecdotes','fichier','tache_text.text','anecdotes_reco','reco_text.text']]
```

```python
EG7_RECO = EG7[['anecdotes_reco', 'fichier', 'reco_text.text']].loc[35:66]
EG7_RECO = EG7_RECO.reset_index()
EG7_RECO['reco_text.text'] = EG7_RECO['reco_text.text'].str.upper()
EG7_RECO['reco_text.text'] = EG7_RECO['reco_text.text'].replace(
    'DONNÉES',
    'DONNEES').replace('FAIT', 'FAITS').replace('DONNEE', 'DONNEES').replace(
        'POINT',
        'POINTS').replace('ITEM', 'ITEMS').replace('NOM', 'NOMS').replace(
            'INFO', 'INFOS').replace('DONNÉE', 'DONNEES').replace(
                'DONNÉS', 'DONNEES').replace('DONNÉ', 'DONNEES').replace(
                    'POINT\n', 'POINTS').replace('FAIT\n', 'FAITS').replace('ITEMS\n','ITEMS')
EG7_RECO
```
