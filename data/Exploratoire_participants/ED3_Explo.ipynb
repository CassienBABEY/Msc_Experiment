```python
ED3 = pd.read_csv("data_expe/ED3_data.csv")
ED3
```

```python
ED3 = ED3[['anecdotes','fichier','tache_text.text','anecdotes_reco','reco_text.text']]
```

```python
ED3_RECO = ED3[['anecdotes_reco', 'fichier', 'reco_text.text']].loc[35:66]
ED3_RECO = ED3_RECO.reset_index()
ED3_RECO['reco_text.text'] = ED3_RECO['reco_text.text'].str.upper()
ED3_RECO['reco_text.text'] = ED3_RECO['reco_text.text'].replace(
    'DONNÉES',
    'DONNEES').replace('FAIT', 'FAITS').replace('DONNEE', 'DONNEES').replace(
        'POINT',
        'POINTS').replace('ITEM', 'ITEMS').replace('NOM', 'NOMS').replace(
            'INFO', 'INFOS').replace('DONNÉE',
                                     'DONNEES').replace('DONNÉS', 'DONNEES').replace('DONNÉ','DONNEES')
ED3_RECO
```
