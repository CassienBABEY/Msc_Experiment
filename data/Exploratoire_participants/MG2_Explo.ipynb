```python
import pandas as pd
import numpy as np
```

```python
MG2 = pd.read_csv("data_expe/MG2_data.csv")
MG2
```

```python
MG2 = MG2[['anecdotes','fichier','tache_text.text','anecdotes_reco','reco_text.text']]
```

```python
MG2_RECO = MG2[['anecdotes_reco','fichier','reco_text.text']].loc[35:66]
MG2_RECO = MG2_RECO.reset_index()
MG2_RECO['reco_text.text'] = MG2_RECO['reco_text.text'].str.upper()
MG2_RECO['reco_text.text'] = MG2_RECO['reco_text.text'].replace(
    'DONNÉES',
    'DONNEES').replace('FAIT', 'FAITS').replace('DONNEE', 'DONNEES').replace(
        'POINT',
        'POINTS').replace('ITEM', 'ITEMS').replace('NOM', 'NOMS').replace(
            'INFO', 'INFOS').replace('DONNÉE',
                                     'DONNEES').replace('DONNÉS', 'DONNEES').replace('DONNÉ','DONNEES')
MG2_RECO
```
