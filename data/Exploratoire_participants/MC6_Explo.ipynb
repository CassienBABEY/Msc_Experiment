```python
import pandas as pd
import numpy as np
```

```python
MC6 = pd.read_csv("data_expe/MC6_data.csv")
MC6
```

```python
MC6 = MC6[['anecdotes','fichier','tache_text.text','anecdotes_reco','reco_text.text']]
```

```python
MC6_RECO = MC6[['anecdotes_reco', 'fichier', 'reco_text.text']].loc[35:66]
MC6_RECO = MC6_RECO.reset_index()
MC6_RECO['reco_text.text'] = MC6_RECO['reco_text.text'].str.upper()
MC6_RECO['reco_text.text'] = MC6_RECO['reco_text.text'].replace(
    'DONNÉES',
    'DONNEES').replace('FAIT', 'FAITS').replace('DONNEE', 'DONNEES').replace(
        'POINT',
        'POINTS').replace('ITEM', 'ITEMS').replace('NOM', 'NOMS').replace(
            'INFO', 'INFOS').replace('DONNÉE', 'DONNEES').replace(
                'DONNÉS', 'DONNEES').replace('DONNÉ', 'DONNEES').replace(
                    'POINT\n', 'POINTS').replace('FAIT\n', 'FAITS').replace('ITEMS\n','ITEMS')
MC6_RECO
```
