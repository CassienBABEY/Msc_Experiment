```python
import pandas as pd
import numpy as np
```

```python
AP5 = pd.read_csv("data_expe/AP5_data.csv")
AP5
```

```python
AP5 = AP5[['anecdotes','fichier','tache_text.text','anecdotes_reco','reco_text.text']]
```

```python
AP5_RECO = AP5[['anecdotes_reco', 'fichier', 'reco_text.text']].loc[35:66]
AP5_RECO = AP5_RECO.reset_index()
AP5_RECO['reco_text.text'] = AP5_RECO['reco_text.text'].str.upper()
AP5_RECO['reco_text.text'] = AP5_RECO['reco_text.text'].replace(
    'DONNÉES',
    'DONNEES').replace('FAIT', 'FAITS').replace('DONNEE', 'DONNEES').replace(
        'POINT',
        'POINTS').replace('ITEM', 'ITEMS').replace('NOM', 'NOMS').replace(
            'INFO', 'INFOS').replace('DONNÉE', 'DONNEES').replace(
                'DONNÉS', 'DONNEES').replace('DONNÉ', 'DONNEES').replace(
                    'POINT\n', 'POINTS').replace('FAIT\n', 'FAITS').replace('ITEMS\n','ITEMS')
AP5_RECO
```
