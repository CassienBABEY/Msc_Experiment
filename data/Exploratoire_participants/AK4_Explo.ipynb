```python
import pandas as pd
import numpy as np
```

```python
AK4 = pd.read_csv("data_expe/AK4_data.csv")
AK4
```

```python
AK4 = AK4[['anecdotes','fichier','tache_text.text','anecdotes_reco','reco_text.text']]
```

```python
AK4_RECO = AK4[['anecdotes_reco', 'fichier', 'reco_text.text']].loc[35:66]
AK4_RECO = AK4_RECO.reset_index()
AK4_RECO['reco_text.text'] = AK4_RECO['reco_text.text'].str.upper()
AK4_RECO['reco_text.text'] = AK4_RECO['reco_text.text'].replace(
    'DONNÉES',
    'DONNEES').replace('FAIT', 'FAITS').replace('DONNEE', 'DONNEES').replace(
        'POINT',
        'POINTS').replace('ITEM', 'ITEMS').replace('NOM', 'NOMS').replace(
            'INFO', 'INFOS').replace('DONNÉE', 'DONNEES').replace(
                'DONNÉS', 'DONNEES').replace('DONNÉ', 'DONNEES').replace(
                    'POINT\n', 'POINTS').replace('FAIT\n', 'FAITS')
AK4_RECO
```
